```python
#Feature tlongest
#the longest actual runtime of the historical jobs of the same class

#groups data by class of job
grouped = df.groupby(["User ID"])

#max value of "Run Time" column for each group
max_runtimes = grouped["Run Time"].max()

#longest actual runtime of historical jobs of same class
longest_runtime = max_runtimes.max()

print(longest_runtime)
```

```python
#Feature tlongest10
#the longest actual runtime of last 10 the historical jobs of the same class

# groups data by class of job
grouped = df.groupby("User ID")

# sort each group by "End Time" and get the last 10 jobs
last_ten_jobs = grouped.apply(lambda x: x.nlargest(10, "End Time"))

# get the maximum "Run Time" of the last 10 jobs for each class
max_runtimes = last_ten_jobs.groupby("User ID")["Run Time"].max()

# get the longest actual runtime of historical jobs of same class
longest_runtime = max_runtimes.max()

print(longest_runtime)
```
